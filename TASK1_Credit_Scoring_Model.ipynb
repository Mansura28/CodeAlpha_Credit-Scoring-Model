## **Import libraries**

```python
import pandas as pd
import numpy as np
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
```

## **Import dataset**

```python
drive.mount('/content/drive')
dataset = pd.read_excel("/content/drive/My Drive/ML projects/Task 1: Credit Scoring Model/Dataset_CreditScoringModel.xlsx")
```

## **Data preparation**

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset = dataset.drop('ID', axis=1)
dataset.shape
```

```python
dataset.isna()
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.fillna(dataset.mean())
dataset.isna().sum()
```

```python
dataset['TARGET'].value_counts()
```

```python
dataset.groupby('TARGET').mean()
```

## **Data split**

```python
y = dataset.iloc[:, 0].values
X = dataset.iloc[:, 1:29].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify = y )
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
joblib.dump(sc, '/content/drive/My Drive/ML projects/Task 1: Credit Scoring Model/Normalized')
```

## **Model (LogisticRegression)**

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
joblib.dump(classifier, '/content/drive/My Drive/ML projects/Task 1: Credit Scoring Model/Classifier')
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

## **Output saving**

```python
predictions = classifier.predict_proba(X_test)
```

```python
df_prediction_prob = pd.DataFrame(predictions, columns = ['prob_0', 'prob_1'])
df_prediction_target = pd.DataFrame(classifier.predict(X_test), columns = ['predicted_TARGET'])
df_test_dataset = pd.DataFrame(y_test,columns= ['Actual Outcome'])

dfx=pd.concat([df_test_dataset, df_prediction_prob, df_prediction_target], axis=1)

dfx.to_csv("/content/drive/My Drive/ML projects/Task 1: Credit Scoring Model/Model_Prediction.xlsx", sep=',', encoding='UTF-8')
dfx.head()
```
